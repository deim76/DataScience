### Задание 1

```python
import pandas as pd
import numpy as np
```

```python
authors=pd.DataFrame({'author_id':[1, 2, 3],'author_name':['Тургенев', 'Чехов', 'Островский']})
authors
```

```python
book=pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                   'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо','Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                   'price':[450, 300, 350, 500, 450, 370, 290]})
book
```

### Задание 2

```python
authors_price=pd.merge(authors,book,on='author_id',how='inner')
authors_price
```

### Задание 3

```python
top5=authors_price.groupby(['author_name','book_title'])['price'].max().tail()
top5
```

### Задание 4

```python
df=authors_price.copy()
df['min_price']=df['price']
df['max_price']=df['price']
df['mean_price']=df['price']
df
```

```python
authors_stat=df.groupby('author_name').agg({'max_price':'max','min_price':'min','mean_price':'mean'})
authors_stat
```
